# Creating and managing experiments
- Provide project name
- Provide experiment name
- For a specific data create a single project
- Inside each project multiple experiments can be created
- Every experiment can be have diferent hyper-parameters attached to it

```
gtf = prototype(verbose=1);
gtf.Prototype("Classification-project-1", "Emergency-vs-Non-Emergency-classifier-experiment-1")
```

# Emergency-vs-Non-Emergency-classifier image classifier - VGG16

# Quick mode training
- Using Default Function
    - dataset_path
    - model_name
    - num_epochs

```
gtf.Default(dataset_path="/content/drive/My Drive/AV_Hack/train_images", 
            model_name="vgg16", 
            num_epochs=5);
```

```
#Start Training
gtf.Train();

#Read the training summary generated once you run the cell and training is completed
```

# Validating the trained classifier

```
gtf = prototype(verbose=1);
gtf.Prototype("Classification-project-1", "Emergency-vs-Non-Emergency-classifier-experiment-1", eval_infer=True);
```


# Load the validation dataset

```
gtf.Dataset_Params(dataset_path="/content/drive/My Drive/AV_Hack/test_images");
gtf.Dataset();
```

# Run validation

```
accuracy, class_based_accuracy = gtf.Evaluate();
```

# Running inference on test images


Load the experiment in inference mode
- Set flag eval_infer as True 

```
gtf = prototype(verbose=1);
gtf.Prototype("Classification-project-1", "Emergency-vs-Non-Emergency-classifier-experiment-1", eval_infer=True)
```

```
gtf.Dataset_Params(dataset_path="/content/drive/My Drive/AV_Hack/Prediction_Images");
gtf.Dataset();
```

# Select image and Run inference

```
import os
predictions = []
for image in os.listdir('/content/drive/My Drive/AV_Hack/Prediction_Images'):
  image_wt_path = '/content/drive/My Drive/AV_Hack/Prediction_Images/' + image
  predictions.append(gtf.Infer(img_name =image_wt_path))
  
```

```
predictions
```

```
preds ={}
for items in predictions:
  image_name = items['img_name'].split("/")[-1]
  predicted_class = items['predicted_class']
  if predicted_class  == 'Emergency_Vehicle':
    labels = 1
  else:
    labels = 0
  preds[image_name] =labels
```

```
preds
```

```
import pandas as pd
predict_label = pd.read_csv('/content/drive/My Drive/AV_Hack/test_vc2kHdQ.csv')
predict_label['emergency_or_not'] = predict_label['image_names'].map(preds)
```

## Import the predictions made
